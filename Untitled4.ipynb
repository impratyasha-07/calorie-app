```python
from google.colab import files
uploaded = files.upload()
import pandas as pd
df = pd.read_csv("calories (1).csv")
df.head()
df.info()
df.isnull().sum()
df.duplicated().sum()
df.drop_duplicates(inplace=True)
df.drop("User_ID", axis=1, inplace=True)

cols_to_convert = ['Age', 'Height', 'Weight', 'Duration', 'Heart_Rate', 'Body_Temp', 'Calories']

for col in cols_to_convert:
    df[col] = pd.to_numeric(df[col], errors='coerce')
df.dropna(subset=cols_to_convert, inplace=True)
df['Gender'] = df['Gender'].str.strip().str.lower()
df['Gender'] = df['Gender'].map({'male': 0, 'female': 1})
# First, convert height from cm to meters
df['Height_m'] = df['Height'] / 100

# Calculate BMI
df['BMI'] = df['Weight'] / (df['Height_m'] ** 2)

# Optional: Drop the intermediate 'Height_m' column
df.drop(columns=['Height_m'], inplace=True)

df.info()
df.describe()

import seaborn as sns
import matplotlib.pyplot as plt

for col in ['Age', 'Weight', 'Duration', 'Calories']:
    sns.histplot(df[col], kde=True)
    plt.title(f"Distribution of {col}")
    plt.show()
sns.boxplot(data=df[['Age', 'Weight', 'Duration', 'Calories']])
plt.title("Outlier Detection")
plt.show()
df = df[df['Calories'] <= 250]
df = df[df['Duration'] <= 45]
sns.heatmap(df.corr(), annot=True, cmap='coolwarm', fmt=".2f")
plt.title("Correlation Matrix")
plt.show()
sns.boxplot(x='Gender', y='Calories', data=df)
plt.title("Calories Burned by Gender")
plt.show()
sns.histplot(df['BMI'], kde=True)
plt.title("Distribution of BMI")
plt.show()
sns.heatmap(df.corr(), annot=True, cmap='coolwarm', fmt=".2f")
plt.title("Updated Correlation Matrix")
plt.show()
sns.scatterplot(x='BMI', y='Calories', data=df)
plt.title("Calories Burned vs BMI")
plt.show()
df.to_csv("calories_cleaned.csv", index=False)

def bmi_category(bmi):
    if bmi < 18.5:
        return 'Underweight'
    elif bmi < 25:
        return 'Normal'
    elif bmi < 30:
        return 'Overweight'
    else:
        return 'Obese'

df['BMI_Category'] = df['BMI'].apply(bmi_category)
import seaborn as sns
import matplotlib.pyplot as plt

sns.boxplot(x='BMI_Category', y='Calories', data=df)
plt.title("Calories Burned by BMI Category")
plt.show()
sns.boxplot(x='Gender', y='Calories', data=df)
plt.title("Calories Burned by Gender")
plt.show()
sns.scatterplot(x='Duration', y='Calories', hue='Gender', data=df)
plt.title("Duration vs Calories by Gender")
plt.show()
bins = [0, 18, 30, 45, 60, 100]
labels = ['<18', '1830', '3145', '4660', '60+']
df['Age_Group'] = pd.cut(df['Age'], bins=bins, labels=labels)
sns.boxplot(x='Age_Group', y='Calories', data=df)
plt.title("Calories Burned by Age Group")
plt.show()
sns.lmplot(x='Duration', y='Calories', hue='BMI_Category', data=df, aspect=1.5)
plt.title("Calories vs Duration by BMI Category")
plt.show()
sns.pairplot(df[['Calories', 'Weight', 'Duration', 'BMI']], diag_kind='kde')
group_stats = df.groupby(['BMI_Category', 'Gender'])[['Calories', 'Duration']].mean().round(1)
print(group_stats)  # Optional, for console output
group_stats.to_csv("grouped_summary.csv")

numeric_df = df.select_dtypes(include='number')  # Only keeps numeric columns
import seaborn as sns
import matplotlib.pyplot as plt

plt.figure(figsize=(10, 6))
sns.heatmap(numeric_df.corr(), annot=True, cmap='coolwarm', fmt=".2f")
plt.title("Correlation Matrix (Numeric Features Only)")
plt.show()


df.to_csv("calories_advanced_eda.csv", index=False)



```
